```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from itertools import combinations
import pandas as pd


# Function to calculate Levenshtein distance between two strings
def levenshtein_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for index2, char2 in enumerate(s2):
        new_distances = [index2 + 1]
        for index1, char1 in enumerate(s1):
            if char1 == char2:
                new_distances.append(distances[index1])
            else:
                new_distances.append(1 + min((distances[index1], distances[index1 + 1], new_distances[-1])))
        distances = new_distances

    return distances[-1]

# Create a dictionary to store similar models for each company
similar_models = {}

# Iterate through each unique company
unique_companies = data['CompanyName'].unique()
for company in unique_companies:
    # Get models for the current company
    company_models = data[data['CompanyName'] == company]['ModelName'].unique().tolist()

    # Find similar models within the company
    similar_models[company] = {}
    for model1, model2 in combinations(company_models, 2):
        distance = levenshtein_distance(model1, model2)
        similarity = 1 - (distance / max(len(model1), len(model2)))
        if similarity > 0.8:  # You can adjust the threshold as needed
            similar_models[company].setdefault(model1, []).append(model2)
            similar_models[company].setdefault(model2, []).append(model1)

# Display the results
for company, similar_models_dict in similar_models.items():
    print(f"Similar models within {company}:")
    for model, similar_models_list in similar_models_dict.items():
        print(f"  {model}: {', '.join(similar_models_list)}")
    print("\n")
```

```python
# Custom function to process each model name
def process_model_name(model_name):
    # Split the model name
    parts = model_name.split()
    # print(parts)
    
    # Check for floating point numbers in each part
    modified_parts = []
    for part in parts:
        if '.' in part and part.replace('.', '').isdigit():
            continue
        else:
            # If not a float, keep the part in the modified list
            modified_parts.append(part)
    
    # Join the modified parts to get the modified string
    modified_string = ' '.join(modified_parts)
    
    # If the string was modified, print the original and modified strings
    if modified_string != model_name:
        print(f"Original: {model_name}, Modified: {modified_string}")
    
    return modified_string

# Apply the custom function to the 'ModelName' column
data['ModelName'] = data['ModelName'].apply(process_model_name)

# Display the DataFrame with modified 'ModelName' column
# print(data)
```

```python
def frequency_encode(series):
    return series.apply(lambda x: series.value_counts(normalize=True).get(x, 0))

# Define the column transformer
encoder_transformers = [
    ('freq_encode_company', FunctionTransformer(frequency_encode, validate=False), ['CompanyName']),
    ('freq_encode_model', FunctionTransformer(frequency_encode, validate=False), ['ModelName']),
    ('one_hot_encode_column0', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [0]),
    ('one_hot_encode_column1', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [1])
]

column_transformer = ColumnTransformer(transformers=encoder_transformers, remainder='passthrough')

# Define the model
model = LinearRegression()

# Create the pipeline
pipeline = Pipeline([
    ('encoding', column_transformer),
    ('model', model)
])

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train)

# Make predictions on the test data
y_pred = pipeline.predict(X_test)

# Evaluate the model
print('R2 Score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
